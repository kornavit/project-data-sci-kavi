## อธิบายถึงข้อมูล
Kiva.org เป็นแพลตฟอร์มการระดมทุนออนไลน์เพื่อขยายบริการทางการเงินให้กับคนยากจนและถูกกีดกันทางการเงินทั่วโลก ผู้ให้กู้ Kiva ได้ให้เงินกู้ยืมมากกว่า 1 พันล้านดอลลาร์แก่ผู้คนกว่า 2 ล้านคน เพื่อกำหนดลำดับความสำคัญของการลงทุน ช่วยแจ้งผู้ให้กู้ และทำความเข้าใจกับชุมชนเป้าหมาย การทราบระดับความยากจนของผู้กู้แต่ละรายเป็นสิ่งสำคัญ อย่างไรก็ตาม ต้องใช้การอนุมานตามชุดข้อมูลที่จำกัดสำหรับผู้กู้แต่ละราย

## รายละเอียดแต่ละ columns
* id: เลขประจำตัวของสัญญากู้ยืมเงิน (loan) บนแพลตฟอร์ม Kiva
* funded_amount: จำนวนเงินที่ได้รับการสนับสนุน (funding) จากผู้ให้กู้ในสัญญานี้
* loan_amount: จำนวนเงินที่ถูกขอกู้ในสัญญานี้
* activity: กิจกรรมหลักที่ผู้กู้ต้องการใช้เงินกู้ในการดำเนินชีวิต
* sector: ภูมิภาคของกิจกรรมหลัก (เช่น การศึกษา, การผลิต, การค้าส่ง)
* use: วัตถุประสงค์หลักที่ผู้กู้ต้องการใช้เงินกู้
* country_code: รหัสประเทศของที่อยู่ของผู้กู้
* country: ชื่อประเทศของที่อยู่ของผู้กู้
* region: ชื่อของเมืองหรือชื่อภูมิภาค
* currency: สกุลเงินที่ใช้ในสัญญานี้
* partner_id: รหัสพันธมิตรที่เชื่อมโยงกับสัญญานี้ (หน่วยงานที่ช่วยสนับสนุนการกู้ยืมเงินใน Kiva)
* posted_time: วันที่และเวลาที่สัญญาถูกโพสต์บนแพลตฟอร์ม Kiva
* disbursed_time: วันที่และเวลาที่เงินกู้
* funded_time: วันที่และเวลาที่เงินกู้ถูกเตมเตมจนสมบรณบนแพลตฟอรม Kiva
* term_in_months: ระยะเวลาที่ต้องชำระต่อเดือน
* lender_count: จำนวนผู้ให้กู้ที่มีส่วนร่วมในการสนับสนุนสัญญาที่บัญชาใน 1 การกู้
* tags: แทกทเกยวของกบสญญาน (คำอธบายยอยเกยวกบการใชเงนก)
* borrower_genders: เพศของผู้กู้
* repayment_interval: รูปแบบการชำระเงินกู้ (รายเดือน, รายปี, ไม่ระบุ)
* date: วันที่และเวลาที่สัญญาถูกนำเข้ามายงฐานขอมลของ Kiva

# เอาค่ามาเปลี่ยนเป็น datetime ใน columns date

```python
df['date'] = pd.to_datetime(df['date'])
```

```python
df['count'] = 1
```

# เช็ดว่ามี columns ไหนเป็น null บ้าง

```python
df.isnull().sum()
```

# คัดเอาแค่ row ที่ไม่มี null

```python
df = df.loc[(df['id'].notna()) &
        (df['country'].notna()) &
        (df['region'].notna()) &
        (df['tags'].notna()) ]
```
