## อธิบายถึงข้อมูล
Kiva.org เป็นแพลตฟอร์มการระดมทุนออนไลน์เพื่อขยายบริการทางการเงินให้กับคนยากจนและถูกกีดกันทางการเงินทั่วโลก ผู้ให้กู้ Kiva ได้ให้เงินกู้ยืมมากกว่า 1 พันล้านดอลลาร์แก่ผู้คนกว่า 2 ล้านคน เพื่อกำหนดลำดับความสำคัญของการลงทุน ช่วยแจ้งผู้ให้กู้ และทำความเข้าใจกับชุมชนเป้าหมาย การทราบระดับความยากจนของผู้กู้แต่ละรายเป็นสิ่งสำคัญ อย่างไรก็ตาม ต้องใช้การอนุมานตามชุดข้อมูลที่จำกัดสำหรับผู้กู้แต่ละราย

## รายละเอียดแต่ละ columns ของ kiva mpi region location
* country : ชื่อประเทศที่ภูมิภาคนี้อยู่
* region : ชื่อภูมิภาค
* world_region : ชื่อโซนของโลกที่ภูมิภาคนี้อยู่
* MPI : ค่า MPI ของภูมิภาคนี้
* geo : พิกัดภูมิศาสตร์ของภูมิภาคนี้ (latitude, longitude)
* MPI Rural, MPI Urban : ค่า MPI สำหรับภูมิภาคชนบทและภูมิภาคเมืองของภูมิภาคนี้

## รายละเอียดแต่ละ columns ของ loan theme by region
* Field Partner Name: ชื่อผู้ให้บริการกู้ยืมเงิน
* sector: กลุ่มอุตสาหกรรมหรือกิจกรรมที่ผู้กู้มีส่วนเกี่ยวข้อง เช่น การเกษตร, การผลิต, การค้าปลีก, การศึกษา
* Loan Theme Type : ประเภทของการกู้ยืม เช่น กู้ยืมเพื่อการศึกษา, กู้ยืมเพื่อซื้ออุปกรณ์การผลิต
* country : ชื่อประเทศ
* region : ชื่อภูมิภาค
* amount : จำนวนเงินที่กู้ยืมในหน่วยเงินตราของประเทศนั้นๆ
* funded_amount : จำนวนเงินที่ได้รับการจัดสรรเพื่อการกู้ยืมในหน่วยเงินตราของประเทศนั้นๆ
* repayment_interval : ระยะเวลาในการชำระเงินคืน เช่น ชำระทุกสัปดาห์, ชำระทุกเดือน, ชำระเมื่อได้รับเงิน
* geo : ตำแหน่งภูมิศาสตร์ของภูมิภาคในรูปแบบ latitude, longitude
* rural_pct : บอกสัดส่วนของประชากรในพื้นที่ที่มีลักษณะชนบทหรือไม่เป็นเมือง โดยค่าสูงสุดของตัวแปรนี้จะเท่ากับ 100% หมายความว่าทุกคนในพื้นที่นั้นเป็นประชากรชนบท และค่าต่ำสุดของตัวแปรนี้จะเป็น 0% หมายความว่าทุกคนในพื้นที่นั้นเป็นประชากรในเมือง

## รายละเอียดแต่ละ columns ของ kiva loan

* activity: กิจกรรมหลักที่ผู้กู้ต้องการใช้เงินกู้ในการดำเนินชีวิต
* sector: ภูมิภาคของกิจกรรมหลัก (เช่น การศึกษา, การผลิต, การค้าส่ง)
* use: วัตถุประสงค์หลักที่ผู้กู้ต้องการใช้เงินกู้
* country_code: รหัสประเทศของที่อยู่ของผู้กู้

* region: ชื่อของเมืองหรือชื่อภูมิภาค
* currency: สกุลเงินที่ใช้ในสัญญานี้

* posted_time: วันที่และเวลาที่สัญญาถูกโพสต์บนแพลตฟอร์ม Kiva
* disbursed_time: วันที่และเวลาที่เงินกู้
* funded_time: วันที่และเวลาที่เงินกู้ถูกเตมเตมจนสมบรณบนแพลตฟอรม Kiva
* term_in_months: ระยะเวลาที่ต้องชำระต่อเดือน
* lender_count: จำนวนผู้ให้กู้ที่มีส่วนร่วมในการสนับสนุนสัญญาที่บัญชาใน 1 การกู้
* tags: แทกทเกยวของกบสญญาน (คำอธบายยอยเกยวกบการใชเงนก)
* borrower_genders: เพศของผู้กู้
* repayment_interval: รูปแบบการชำระเงินกู้ (รายเดือน, รายปี, ไม่ระบุ)
* date: วันที่และเวลาที่สัญญาถูกนำเข้ามายงฐานขอมลของ Kiva

# ทำความสะอาดของข้อมูล (DATA CLEANING)
ทำการรวมข้อมูลที่ต้องการใช้งาน นั่นคือ เอา kiva_mpi_region_locations.csv และ loan_themes_by_region.csv มารวมกันเพื่อให้เราสามารถวิเคราะห์ความสัมพันธ์ระหว่างการกู้ยืมและสภาพเศรษฐกิจและสังคมของแต่ละภูมิภาคหรือพื้นที่ในแต่ละประเทศได้ โดยทั้งสองตารางนี้มีข้อมูลเกี่ยวกับพื้นที่ของแต่ละประเทศ โดย kiva_mpi_region_locations.csv มีข้อมูลเกี่ยวกับดัชนีความยากจนในแต่ละพื้นที่ของแต่ละประเทศ และ loan_themes_by_region.csv มีข้อมูลเกี่ยวกับประเภทของการกู้ยืมและจำนวนเงินที่กู้ยืมในแต่ละพื้นที่ของแต่ละประเทศ

```python
import pandas as pd
import numpy as np
import seaborn as sns
%pylab inline
```

```python
# อ่านไฟล์ csv และเก็บข้อมูลใน dataframe
df_kiva_mpi = pd.read_csv('dataset/kiva_mpi_region_locations.csv')
kiva_loans = pd.read_csv('dataset/kiva_loans.csv')
df_loan_themes_by_region = pd.read_csv('dataset/loan_themes_by_region.csv')
```

# Clean Data kiva_loans.csv
ลบคอลัมน์ที่ไม่ต้องการ

```python
kiva_loans = kiva_loans.drop(['use', 'tags', 'borrower_genders'], axis=1)
```

แปลงคอลัมน์ funded_time, planned_expiration_time, posted_time เป็น datetime format

```python
kiva_loans['funded_time'] = pd.to_datetime(kiva_loans['funded_time'])
kiva_loans['posted_time'] = pd.to_datetime(kiva_loans['posted_time'])
kiva_loans['disbursed_time'] = pd.to_datetime(kiva_loans['disbursed_time'])
```

แปลงคอลัมน์ term_in_months เป็น float และ ปัดเศษลง

```python
kiva_loans['term_in_months'] = kiva_loans['term_in_months'].astype(float).apply(np.floor)
```

แปลงคอลัมน์ loan_amount เป็น int

```python
kiva_loans['loan_amount'] = kiva_loans['loan_amount'].astype(int)
```

แปลงคอลัมน์ lender_count เป็น int

```python
kiva_loans['lender_count'] = kiva_loans['lender_count'].astype(int)
```

แปลงคอลัมน์ country_code เป็น category

```python
kiva_loans['country_code'] = kiva_loans['country_code'].astype('category')
```

แปลงคอลัมน์ repayment_interval เป็น category

```python
kiva_loans['repayment_interval'] = kiva_loans['repayment_interval'].astype('category')
```

ลบแถวที่มี missing value

```python
kiva_loans = kiva_loans.dropna()
```

ลบแถวที่ loan_amount เป็น 0

```python
kiva_loans = kiva_loans[kiva_loans['loan_amount'] != 0]
```

# Clean Data loan_themes_by_region.csv
แทนค่า missing value ใน column amount ด้วยค่าเฉลี่ยของ column นั้น

```python
amount_mean = df_loan_themes_by_region["amount"].mean()
df_loan_themes_by_region["amount"].fillna(amount_mean, inplace=True)
```

# Clean kiva_mpi_region_locations.csv
เลือกและเปลี่ยนชื่อ columns ที่ต้องการ

```python
df_kiva_mpi=df_kiva_mpi[['LocationName', 'ISO', 'country', 'region', 'world_region', 'MPI', 'geo', 'lat', 'lon']]
df_kiva_mpi.columns = ['location_name', 'iso_code', 'country', 'region', 'world_region', 'mpi', 'geo', 'latitude', 'longitude']
```

# แปลงตัวเลขจาก string เป็น float

```python
df_kiva_mpi['mpi'] = df_kiva_mpi['mpi'].astype(float)
```

ลบแถวที่มีค่า MPI เป็น NaN

```python
df_kiva_mpi=df_kiva_mpi.dropna(subset=['mpi'])
```

เมื่อทำ การ clean data ทั้งสาม dataset แล้ว เราจะนำมารวมกันโดยการ merge เพื่อนำไปวิเคราะห์ และ หา insight ขอ ข้อมูลที่น่าสนใจต่อไป

```python
df_merged = pd.merge(kiva_loans,df_loan_themes_by_region, on='region', how='inner')
df_merged
```

```python
df_merged = pd.merge(df_merged, df_kiva_mpi, on='region', how='inner')
df_merged
```

## สำรวจแต่ละประเทศที่มีการกู้ยืมว่าจ่ายเป็นรูปแบบใดบ้าง

```python
df_loans = pd.read_csv('kiva_loans.csv', usecols=['country', 'repayment_interval'])
repayment_counts = df_merged.groupby(['country', 'repayment_interval']).size().reset_index(name='counts')
repayment_totals = df_merged.groupby('country').size().reset_index(name='totals')
repayment_summary = pd.merge(repayment_counts, repayment_totals, on='country')
repayment_summary['proportions'] = repayment_summary['counts'] / repayment_summary['totals']

repayment_summary
```

```python
sns.set_style('whitegrid')
g = sns.catplot(data=repayment_summary, x='country', y='proportions', hue='repayment_interval', kind='bar', height=5, aspect=2)
g.despine(left=True)
g.set_ylabels('Proportion of loans')
g.set_xlabels('Country')
plt.xticks(rotation=90)
plt.show()
```

```python
mpi = pd.read_csv('kiva_mpi_region_locations.csv')
mpi = mpi.dropna(how='any')
mpi
```

ปรับตารางให้เข้ากับข้อมูลที่เราต้องการในที่นี้คือ ประเทศ แล้ว ภูมิภาค

```python
mpi = mpi.reset_index()
mpi = mpi.set_index(['country','region'])
mpi
```

```python
clear = clear.reset_index()
clear = clear.set_index(['country','region'])
clear
```

ทำการต่อ dataset 2 อัน เพื่อเชื่อมข้อมูล

```python
clear = clear.merge(mpi, on = ['country','region'], how = 'left')
clear = clear.dropna(how='any')
clear
```

ทำการหาค่าเฉลี่ยของ MPI เพราะในแต่ละภูมิภาคไม่ได้มีบันทึกแค่ค่า ๆเดียว

```python
MPI_mean = clear.groupby(['country','region']).mean()
MPI_mean = MPI_mean.reset_index()
MPI_mean = MPI_mean.loc[:,['country','region','MPI']]
MPI_mean = MPI_mean.set_index(['country','region'])
MPI_mean
```

เคลียร์ column เก่าทิ้งเพื่อนำค่า MPI ที่หามาเชื่อมใหม่

```python
clear_sum = clear.groupby(['country','region']).sum()
clear_sum = clear_sum.drop(columns=['MPI'])
clear_sum
```

ต่อตารางเพื่อเชื่อมค่า MPI กับภูมิภาคนั้น ๆ

```python
clear_sum = clear_sum.join(MPI_mean, on = ['country','region'], how = 'left')
clear_sum
```

ตัดข้อมูลที่ต้องการมาแสดง

```python
clear = clear_sum.reset_index().loc[:,['country','region','loan_amount','MPI']]
clear
```

หาจำนวนเงินทั้งหมดของการกู้ยืม

```python
sumloan = clear.loan_amount.sum()
sumloan
```

เทียบ percent การให้กู้ยืมเงินทั้งหมดในแต่ละภูมิภาค

```python
clear['percent_loan'] = clear.loan_amount / sumloan * 100 
# clear.loc[:,['MPI','percent_loan']]
clear = clear.reset_index() 
clear['ct_code'] = clear.country.astype(str).str[0] + clear.region.astype(str).str[0] + clear.index.astype(str)
clear = clear.set_index('ct_code')
clear
```

เปรียบเทียบค่าออกมาเป็น dataframe สุดท้าย

```python
clear = clear.reset_index()
clear = clear.loc[:,['ct_code','MPI','percent_loan']]
clear

clear1 = clear.loc[0:10,:]
clear1
```

# สาเหตุหลักที่ทำให้กู้ยืมไม่สามารถชำระเงินคืนได้คืออะไรในแต่ละประเทศ

```python
import matplotlib.pyplot as plt
# กรองข้อมูลเฉพาะที่ไม่สามารถชำระเงินคืนได้
df_defaulted = df_merged[df_merged['repayment_interval'] == 'irregular'].copy()
df_defaulted
```

```python
# หาสาเหตุหลักที่ทำให้กู้ยืมไม่สามารถชำระเงินคืนได้ในแต่ละประเทศ
grouped = df_defaulted.groupby('country')['id'].count().reset_index()
grouped = grouped.rename(columns={'id': 'irregular_loans'})
grouped = grouped.sort_values('irregular_loans', ascending=False)
grouped
```

```python
# พล็อตกราฟแสดงจำนวนกู้ยืมที่ไม่สามารถชำระเงินคืนได้ในแต่ละประเทศ
plt.bar(grouped['country'], grouped['irregular_loans'])
plt.xticks(rotation=90)
plt.xlabel('Country')
plt.ylabel('Number of irregular loans')
plt.title('Top Countries with Highest Number of irregular Loans')
plt.show()
```

จากกราฟที่ได้ จะสามารถดูสาเหตุหลักที่ทำให้กู้ยืมไม่สามารถชำระเงินคืนได้ในแต่ละประเทศ ได้โดยดูจากจำนวนผู้กู้ยืมที่ไม่สามารถชำระเงินคืนได้ (default) ของแต่ละประเทศ และเปรียบเทียบกับจำนวนผู้กู้ยืมทั้งหมดในแต่ละประเทศ ว่าสัดส่วนเป็นอย่างไร หากสัดส่วนผู้กู้ยืมที่ไม่สามารถชำระเงินคืนได้สูงกว่าประเทศอื่น ๆ แสดงว่าสาเหตุหลักอยู่ที่ประเทศนั้น ๆ ได้ซึ่งอาจจะเกี่ยวกับสภาพเศรษฐกิจและสังคม หรือเกี่ยวกับนโยบายการกู้ยืมที่ไม่เหมาะสมหรือไม่คำนึงถึงความสามารถในการชำระเงินของผู้กู้ยืมในประเทศนั้น ๆ
